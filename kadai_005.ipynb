```python
# 台形の面積
# 変数に数値を代入
width1 = 10
width2 = 20
length = 5


# 計算結果を変数に代入
area = (width1 + width2) * length / 2

# f文字列を使って計算結果と単位（cm²）を表示
print(f"面積は {area} cm² です。")

```
